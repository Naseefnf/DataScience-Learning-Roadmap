```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
final = pd.merge(df, df_country, on='Country Code', how='left')
final
```

```python
final.dtypes
```

```python
final.isnull().sum()
```

```python
final.Country.value_counts()
```

```python
country_names = final.Country.value_counts().index
country_names
```

```python
country_count = final.Country.value_counts().values
country_count
```

```python
# top 3 countries that use zomato
plt.pie(country_count[:3], labels=country_names[:3], autopct="%.3f%%")
plt.show()
```

Zomato's maximum sales are from India followed by USA and UK

```python
ratings = final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating count"})
ratings
```

## Observations
1. When rating between 4.5 to 4.9 is excellent
2. From 4.0 to 4.4 rating is very good
3. When rating between 3.5 to 3.9 rating is good
4. When rating between 2.5 to 3.4 rating is average
5. When rating between 1.8 to 2.4 rating is poor

```python
ratings.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings, palette='hls')
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y='Rating count',hue='Rating color', data=ratings, palette=['white','red','orange','yellow','green','green'])
```

## Obeservation
1. Not rated count is vary high
2. Maximum number of rating is between 2.5 to 3.5

```python
sns.countplot(x=ratings['Rating color'], palette=['white','red','orange','yellow','green','green'])
```

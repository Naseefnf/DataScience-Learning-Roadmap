```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Univariate Visualizations

1. Histogram + KDE

```python
sns.histplot(tips['tip'], kde=True, bins=20)
```

2. Pure KDE Plot:

```python
sns.kdeplot(tips['tip'])
```

3. Box Plot:

```python
sns.boxplot(y=tips['tip'])
```

4. voilin plot:

```python
sns.violinplot(y=tips['tip'])
```

5. Rug Plot:

```python
sns.rugplot(tips['tip'])
```

6. Countplot

```python
sns.countplot(x='day', data=tips, palette='pastel')
```

```python
sns.countplot(x='sex', data=tips)
```

### Bivariate Visualization

1. Scatter plot with regression

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

Adding linear fit

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

2. jointplot (combo of scatter, hist, KDE):

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

3. Pairplot

```python
sns.pairplot(tips, hue='sex', palette='Blues')
```

```python
sns.pairplot(tips, hue='sex', palette='Blues', kind='reg')
```

4. Swarmplot

```python
sns.swarmplot(x='day', y='total_bill',hue='sex', data=tips, color='Blue')
```

5. Stripplot

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True,hue='sex', dodge=True)
#jitter=True adds random noise to reduce overlap.
#dodge=True separates hue categories side-by-side.
```

### Multivariate Visualization

1. Heatmap

```python
corr_matrix = tips.corr(numeric_only=True)
sns.heatmap(corr_matrix,annot=True)
```

```python
corr_matrix = tips.corr(numeric_only=True)
sns.heatmap(corr_matrix,annot=True, cmap='coolwarm')
#cmap='coolwarm' colors positive with warm and negative with cool colors.
#annot=True displays correlation values on the heatmap.
```

2. cluster map

```python
sns.clustermap(corr_matrix, cmap='viridis', standard_scale=1)
```

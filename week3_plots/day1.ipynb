## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5]
y = [34,33,28,36,32]
```

```python
plt.plot(x,y, color='purple', linestyle='-.', linewidth='0.5', marker='*')
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")
```

```python
plt.plot(x, y, 'rD--', markersize=4, alpha=0.2 )
```

```python
day = np.arange(1,31)
low_temp = np.random.randint(30,35,30)
high_temp = np.random.randint(34,39,30)
avg_temp = (low_temp + high_temp) / 2
```

```python
plt.plot(day, low_temp, label="low")
plt.plot(day, high_temp, label="high")
plt.plot(day, avg_temp, label="avg")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.legend(loc="best", shadow=True, fontsize="small")
plt.grid(color="black", alpha=0.4)

```

Bar chart

```python
fruits = ['Apple', 'Banana', 'Cherry', 'Grapes', 'Orange']
sales = [50, 80, 40, 75, 60]
profit = [55,88,38,79,46]
```

```python
xpos = np.arange(len(fruits))
```

```python
plt.xticks(xpos, fruits)
plt.bar(xpos-0.2, sales, width=0.4, label="sales")
plt.bar(xpos+0.2, profit, width=0.4, label="profit")
plt.xlabel("Fruits", color="red")
plt.ylabel("Sales", color="red")
plt.legend()
```

```python
plt.yticks(xpos, fruits)
plt.barh(xpos-0.2, sales, height=0.4, label="sales")
plt.barh(xpos+0.2, profit, height=0.4, label="profit")
plt.legend()
```

Histograms

```python
np.random.seed(0)
blood_sugar_men = np.random.randint(60,130,10)
blood_sugar_women = np.random.randint(60,130,10)
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=4, rwidth=0.95, label=['Men','Women'])
plt.legend(fontsize='small')
plt.xlabel("Sugar Level")
plt.ylabel("No. of persons")
```

```python
plt.hist([blood_sugar_men, blood_sugar_women],
          bins=4, rwidth=0.95, 
          label=['Men','Women'],
          orientation='horizontal')
plt.legend(fontsize='small')
plt.ylabel("Sugar Level")
plt.xlabel("No. of persons")
```

Pie chart

```python
expense_categories = ['Rent', 'Groceries', 'Utilities', 'Transport', 'Entertainment', 'Savings', 'Other']
monthly_expenses = [15000, 4000, 2000, 1800, 1200, 3000, 1000]
```

```python
plt.pie(monthly_expenses, labels=expense_categories)
plt.axis('equal')
plt.show()
```

```python
plt.pie(monthly_expenses, labels=expense_categories, autopct='%0.2f%%')
plt.axis('equal')
plt.show()
```

```python
plt.pie(monthly_expenses, labels=expense_categories, autopct='%0.2f%%', explode=[0,0.05,0,0,0,0,0])
plt.axis('equal')
plt.show()
```

```python
plt.pie(monthly_expenses, labels=expense_categories, 
        autopct='%0.2f%%', 
        explode=[0,0.05,0,0,0,0,0],
        startangle=60)
plt.axis('equal')
plt.show()
plt.savefig('char.png', bbox_inches='tight')
```

Plot a sine and cosine curve on the same chart with labels.

```python
x = np.arange(10,91,10)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='Sine')
plt.plot(x, y_cos, label="Cose")
plt.xlabel("X values")
```

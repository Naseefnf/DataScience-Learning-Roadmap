```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips   ')
tips.head(10)
```

Plot a correlation heatmap of a dataset.


```python
corrmatrix = tips.corr(numeric_only=True)
sns.heatmap(corrmatrix, annot=True, cmap='Reds')
```

Create a cluster map to group similar rows/columns.


```python
sns.clustermap(data=corrmatrix)
```

Customize colormap and add annotations.


```python
sns.heatmap(corrmatrix, cmap='crest', annot=True)
```

Generate a heatmap for a 5Ã—5 random matrix.


```python
import numpy as np

np.random.seed(0)
matrix = np.random.randint(50,100,25).reshape(5,5)
df = pd.DataFrame(matrix)

corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True)
```

Plot correlation heatmap for numeric features in a DataFrame.
Create a heatmap with annotations (values displayed).


```python
np.random.seed(1)
data = pd.DataFrame({
    'A': np.random.randint(0, 100, 10),
    'B': np.random.randint(0, 100, 10),
    'C': np.random.randint(0, 100, 10),
    'D': np.random.randint(0, 100, 10),
    'E': np.random.randint(0, 100, 10)
})

df = pd.DataFrame(data)
df
```

```python
corr_matrixs = df.corr(numeric_only=True)
sns.heatmap(corr_matrixs, annot=True, cmap='icefire')
```

Customize heatmap with a coolwarm palette.


```python
sns.heatmap(corr_matrixs, annot=True, cmap='coolwarm')
```

Draw a clustermap to group students based on marks in multiple subjects.


```python
df = pd.DataFrame({
    "Student": ["Amit", "Naseef", "Geetha", "Irshad", "Meera"],
    "Maths": [85, 78, 91, 55, 70],
    "Physics": [75, 88, 84, 60, 68],
    "Chemistry": [80, 82, 86, 65, 72],
    "English": [90, 89, 86, 70, 75]
})

df
```

```python
corr = df.corr(numeric_only=True)
sns.clustermap(corr, cmap='Reds')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = [1, 2, 3]
y = [4, 5, 6]
plt.figure(figsize=(10,5))
plt.plot(x, y, marker='*')
plt.title("Simple line plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

Bar charts

```python
categories = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales = [30, 25, 35, 20, 40]
plt.bar(categories, sales)
plt.title('Sales Data')
plt.xlabel('day')
plt.ylabel("Sales")
```

```python
plt.barh(categories,sales)
```

Scatter plots

```python
height = [150, 160, 170, 180, 190]
weight = [50, 60, 70, 80, 90]
plt.scatter(height,weight, marker='*')
plt.xlabel('Height')
plt.ylabel('Weight')
```

 Pie Charts

```python
sizes = [40, 35, 25]
labels = ['Python', 'Java', 'C++']
plt.pie(sizes, labels=labels, 
        autopct='%0.2f%%',
        explode=[0,0.1,0],
        startangle=90
        )
```

Histogram

```python
np.random.seed(0)
sugar_levels = np.random.randint(60, 150, 100)
plt.hist(sugar_levels,bins=5, rwidth=0.5)
```

Box Plot (Box-and-Whisker)

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 15, 100)
plt.boxplot([data1, data2], labels=['Type A', "Type B"])
```

Subplots (Multiple Plots Together)


```python
fig, axes = plt.subplots(1, 4, figsize=(10, 4))
sizes = [40, 35, 25]
labels = ['Python', 'Java', 'C++']
axes[0].hist(data1, bins=10, color='lightgreen')
axes[0].set_title('Type A')
axes[1].pie(sizes, labels=labels)
axes[1].set_title('Type B')
axes[2].hist(data1, bins=10, color='lightgreen')
axes[2].set_title('Type A')
plt.tight_layout(rect=(0.8,0,0,0))
plt.show()
```

Heatmaps (Advanced, If Wanted)

```python
matrix = np.random.rand(5, 5)
plt.imshow(matrix, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title('Random Matrix Heatmap')
plt.show()
```

Plot styling and annotations

```python
height = [150, 160, 170, 180, 190]
weight = [50, 60, 70, 80, 90]
plt.scatter(height,weight, marker='*')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title("Height vs Weight", fontsize=16, fontweight='bold')
```

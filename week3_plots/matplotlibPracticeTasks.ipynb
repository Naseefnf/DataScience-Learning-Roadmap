```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Line Plot

Task: Plot sales growth over months as a line chart. Add markers and a title.



```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [2200, 2700, 1800, 3500, 4000]
plt.plot(months, sales, marker='*')
plt.title("Sales over months", fontweight='bold')
```

#### Bar chart

Task: Show revenue per product using a bar chart. Color bars and add value labels above each bar.

```python
products = ['Pen', 'Book', 'Bag', 'Pencil']
revenue = [320, 510, 390, 410]
chart = plt.bar(products, revenue, color='black')
plt.bar_label(chart, labels=revenue, padding=1)
```

#### Scatter Plot

Task: Plot height vs. weight as a scatter plot. Change marker size or color for fun!



```python
heights = [150, 160, 170, 180, 190]
weights = [50, 55, 70, 80, 95]
plt.scatter(heights, weights, marker='*', c='red')
```

#### Histogram

Task: Show the distribution of exam scores. Use 6 bins. Color the histogram.

```python
exam_scores = np.random.randint(40, 100, 60)
plt.hist(exam_scores, bins=6, rwidth=0.9)
```

#### Box plot

Task: Compare math and science scores using a box plot. Add grid lines.



```python
math_scores = [80, 85, 78, 90, 88, 91, 67, 76, 89, 95]
science_scores = [75, 82, 77, 92, 85, 90, 70, 74, 84, 93]
english_scores = [75, 92, 77, 82, 85, 78, 70, 74, 84, 93]

plt.boxplot([math_scores, science_scores, english_scores])
```

#### Pie Chart


Task: Show market share for each segment as a pie chart. Display percentages and pull out segment 'A'.

```python
segments = ['A', 'B', 'C', 'D']
market_share = [35, 25, 30, 10]
plt.pie(market_share, 
        labels=segments, 
        autopct='%0.2f%%',
        explode=[0.1,0,0,0])
```

#### Subplots


Task: Create a single figure with two subplots: (1st) a bar chart, (2nd) a line chart. Add titles and style each plot differently.

```python
products = ['Pen', 'Book', 'Bag', 'Pencil']
revenue = [320, 510, 390, 410]

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [2200, 2700, 1800, 3500, 4000]

fig, axes = plt.subplots(1,2, figsize=(10,4))
axes[0].bar(products,revenue)
axes[0].set_title("Bar chart")
axes[1].plot(months, sales)
axes[1].plot("Line chart")
```

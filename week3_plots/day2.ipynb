```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

#### Code 

```python
tip = sns.load_dataset('tips')
```

```python
tip.head()
```

Plot a KDE plot for a numeric column.


```python
sns.kdeplot(tip['tip'])
```

Use jointplot to visualize correlation between two variables.


```python
sns.jointplot(x='total_bill', y='tip', data=tip, kind='scatter')
```

Compare two groups with boxplots.


```python
math_scores = [80, 85, 78, 90, 88, 91, 67, 76, 89, 95]
science_scores = [75, 82, 77, 92, 85, 90, 70, 74, 84, 93]

sns.boxplot([math_scores, science_scores])
plt.xticks([0,1], ['Maths', 'Science'])
```
